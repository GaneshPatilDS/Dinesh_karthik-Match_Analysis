```python
D_Kartik_df=df_2024_RCB[df_2024_RCB['batsman']=='KD Karthik']
```

```python
# Kartik Bating against Sunrisers Hyderaba
D_Kartik_df=D_Kartik_df[D_Kartik_df['start_date'] == '2024-04-15']


```

```python
def phase(ball):
    if ball <= 5.6:
        return 'Powerplay'
    elif ball <= 14.6:
        return 'Middle'
    else:
        return 'Death'
```

```python
D_Kartik_df['phase'] = D_Kartik_df['ball'].apply(lambda x: phase(x))
```
