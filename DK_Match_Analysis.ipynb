```python
import math

# Define a function to convert ball to over and ball
def convert_to_over_ball(ball):
    over = math.floor(ball)  # Get the integer part of the ball (the over)
    ball_in_over = round((ball - over) * 10)  # Get the decimal part and multiply by 10 (since there are 6 balls in an over)
    return over, ball_in_over

# Apply the function to the 'ball' column to create 'over' and 'ball' columns
df['over'], df['ball'] = zip(*df['ball'].apply(convert_to_over_ball))
```
